```python
normals = pd.Series(np.random.normal(size=10))
print(normals)
normals.plot()
```

```python
normals.cumsum().plot(grid=True)
```

```python
variables = pd.DataFrame({'normal': np.random.normal(size=15), 
                       'gamma': np.random.gamma(1, size=15), 
                       'poisson': np.random.poisson(size=15)})
print(variables)

variables.cumsum(0).plot()
```

```python
variables.cumsum(0).plot(subplots=True)
```

```python
variables.cumsum(0).plot(secondary_y='normal')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(5, 10))
for i,var in enumerate(['normal','gamma','poisson']):
    variables[var].cumsum(0).plot(ax=axes[i], title=var)
axes[0].set_ylabel('cumulative sum')
```

```python
plt.plot(s, np.sin(s - 0), linestyle='-')        
plt.plot(s, np.sin(s - 1), linestyle=':')           
plt.plot(s, np.sin(s - 2), linestyle='-.')  
```

```python
rng = np.random.RandomState(5)
x = rng.randn(10)
y = rng.randn(10)
colors = rng.rand(10)
sizes = 100 * rng.rand(10)

plt.scatter(x, y, c=colors, s=sizes, alpha=2,
            cmap='coolwarm')
plt.colorbar()  # show color scale
```

```python
x = np.linspace(0, 10,200)
dy = 0.1
y = np.tan(x) + dy * np.random.randn(200)

plt.errorbar(x, y, yerr=dy, fmt='.c');
```
